## Машинное обучение:
Представьте ситуацию, когда вам покакзывают фотографию вашего друга. Несмотря какая бы она не была : затемненная или под разными углами, лысый ваш друг или в рубашке - вы узнаете его. А теперь представим, что у вас задача научить это делать компьютер, как это сделать? 

Давайте опишем стандартный подход к программироавнию: мы передаем компьютеру алгоритм (нашу программу) и данные (имя, возраст пользователя), затем на выходе получаем результат, который жестко описан и детерминирвоан в нашем алгоритме. 
Стандартный подход к программирваонию для решения задачи с классификацией нашего друга нам не подойдет, нам подойдет немного другой подход : мы передаем компьютеру данные(фото друга) и метку(имя нашего друга, то что мы ожидаем на выходе), а компьютер ищет способы, пути для достижения этой цели и формирует алгоритм, который для этого подйодет. Это называется машинным обучением и задача, которая перед нами возникла класической для ML. 

Чтобы выполнить данную задачу, вы можете предоставить компьютеру множетсво примеров (фоток вашего друга, желательно под разными углами) и метку(имя вашего друга) и программа каждый раз будет искать алгоритм, следуя которому она будет приходить к метке и таким алгоритмов будет n, как и ваших фотографий. 
Соответсвенно, чем больше обьем передаваемых данных, тем больше точность вашей системы.  
После этапа обучения ваша программа (система, функция) будет при новых данных, которые она еще не видела, пользоваться схожими алгоритмами,которые она сама же выработала и индифицировать вашего друга на фото, выдавая соотвествующую метку. 
Этот этап называются тестирвоанием, когда нашему алгоритму, который вырабоатал наша система, передаются теестовые данные, на которых система отвечает меткой.

Теперь нам ясна разница, между двумя подходами к программирванию: стандартных подход испоьзуется в ситуациях, когда нам четко ясны правила, по которым мы будем обабатывать входные данные, в то же время машинное обучение позволяет нам не зная точных правил - выработать их. 

На практике ML почти всегда комбинируется с стандартным методом программирования. 
Это понятно, т.к чтобы начать применять методы ML, вам для начала нужно дать понять компьютеру, что он и как он будет это делать - находить, загружать, обрабатывать фотографии (данные). Так же дял облегчения обучения мы будем часто пользоваться методами, способсвующиим увеличению точности, например, эвристические алгоритмы, которые отделяют лица от заднего фона - например от неба, от лате арта и тд.
Так что довольно часто использование комбинирование этих нескольких методов часто приводит к результату, преуспевающему результату отдельно взятого метода.

Алгоритм, в котором прописанно точное действие на какое то событие - будет более предпочтительней и устойчивей к ошибкам, нежели большая модель.

И прочие вещи, где ым можем обойтись без машинного обучения.

Интересно подметить, что ML, в своей сути является математической эмпирической концепцией, которую люди так привыкли применять в

Алгоритмы, точнее их визуализация - сильно упрощенная, но не смотря на это таких функций будет бессконечное множество. Но те не менее у всего этого множетсва будет средняя функция, и нам интуитивно понятон что это будет $f(x) = x$, которая на графике изображана синей линией с круглыми точками.

Чтобы к этому прийти более строго, мы можем представить это в виде формулы: 
$$ n \to \infty $$
$$ f(x)_{med} = \frac{\sum_{i=0}^n{f'(x)_i}}{n} $$


**Простая функция** - функция, которая описывается одной и только одной функцией. Если функция состоит из нескольких функций - она не является простой. Пример непростой функции: $f(x) = f(x)_j + f(x)_{j+1}$

Как мы видим, мы берем сумму всех производных простых функций, которых бессконечное число, и затем делим на **n**, ту же бессконечность. В результате мы должны получить $f(x)_{med} = x$. 
Вопрос, что если мы возьмем множество непростых функций и по тому же методу посчитаем их среднее, будет ли это досиг пор $f(x) = x$ или будет какой-то другой функцией? 



## Доработать 

Тогда мы передаем нашему компьютеру *х* и говорим найти функуцию (а скорее целый набор функций, с набором арифметических операций между ними: 
$$f(x) = {f(x)_1 a f(x)_2 \dots a f(x)_n}$$
$$a = \ \boxed{+,\,-,\,\times,\,\div, \dots}$$


**a** - операнд какого-то рода, представляющий набор арифметических операций (сложение, вычитание, умножение, деление, возведение в степень, корень
**f(x)** - это конечный результат, алгоритм, который возращает нам компьютер, который по факту является ML системой. f(x) на данном этапе отвечает, как минимум одному критерию - сходимости в *x*. Где *x* - это метка, которую мы передали нашей системе.

Понятно, что при первом обучении у нас всегда будет бесконечное количесво функции, которые удовлетворяют, этом условию (приходят в это место *х*). 
Мы скорее всего на данном этапе будем выбирать наиболее оптимальную из всех функции - скорее всего это будет средняя из этих фукнций: 


$$\text{т.е существует бесконечно много } f^{(i)}(x)\ \text{таких, что } f^{(i)}(x) = f(x)$$

И из этих бесконечно многих функций- нам нужно выбрать единое представление, которое скорее всего будет арифметическим средним, т.е: 
$$ N = \infty $$
$$ f^{(I)}(x) = f(x)_j$$
$$ F(x) = \frac{\sum^N_{I=0}{f^{(I)}(x)}}{N} $$ 

Но, понятно, что нас, как программистов с ограниченным ресурсными возможностями, не удовлетворяет значение бессконечности ($\infty$),  которым обладает $N$, т.к мы попросту не может его полностью проанализировать. Поэтому, скорее всего у нас будет выборка из 10, или 100 или куда больше функций, которые максимально близки к среднему арифметичекому. Среднее арифметичсекое из функций мы бы могли записать так: 

У каждой функции, скомбенированной из нескольких функций и операций между ними будет свое среднее, лежащие на базе производных функций, с которых мы собрали имеющийся фунцию: 

$$el_j - \text{обозначаю какой-то элемент из выборки} $$
$$ n - количесвто элменетов в не основной выборке $$
$$ f(x)_j\text{(med)} = \frac{\sum_{i=0}^{n=\infty}{f`_i}(x)}{n} $$

Но интуитивно понятно, что если мы будем иметь бессконечное количесво функций $F(x)$, состоящих из $f(x)$, которые в свою очередь состоят так же из бессконечного количества функций - среднее что для того, что для того, случая будет простое : $F(x) = x$ и $f(x) = x$, соответсвенно. 
Но далеко не всегда $F(x) = x$ и $f(x) = x$ - будет тем что нам нужно, тем самым алгоритмом, который подойдет для конкретного случая, поэтому у нас будет целый набор функций $F(x)_j$ и $f(x)_j$, приближенных к их среднему, или же взяты рандомно, этот вопрос следует исследовать. Но если мы возьмеме их рандомно, мы по факту из бессконечного количсева алгоритмов выберем **n** рандомных и тестируя определенный случай, мы сможем выбирать из них наиболее подходящие, которые менее ошибочные, следовательно узнавая область пространсва из все возможных, помещенных в 90 градусов оси [х;y], которую мы можем описывать, к примеру двумя векторами, которые имееют определнный угол между собой.
